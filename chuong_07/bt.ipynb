```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk

root = tk.Tk()

path = filedialog.askopenfilename(
    filetypes=[("Image files", "*.png *.jpg *.jpeg *.gif")]
)

if path:
    img = ImageTk.PhotoImage(Image.open(path))
    tk.Label(root, image=img).pack()

root.mainloop()
```

```python
import tkinter as tk

def divisors():
    n = int(e.get())
    d = [str(i) for i in range(1, n+1) if n % i == 0]
    lbl.config(text=" ".join(d))

root = tk.Tk()
e = tk.Entry(root)
e.pack()
tk.Button(root, text="Validate", command=divisors).pack()
lbl = tk.Label(root)
lbl.pack()
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show():
    name = e1.get()
    age = int(e2.get())
    msg = f"Xin chào {name}\n"
    msg += "Bạn đã trưởng thành" if age >= 18 else "Bạn còn nhỏ tuổi"
    messagebox.showinfo("Thông báo", msg)

root = tk.Tk()
tk.Label(root, text="Name").pack()
e1 = tk.Entry(root); e1.pack()
tk.Label(root, text="Age").pack()
e2 = tk.Entry(root); e2.pack()
tk.Button(root, text="OK", command=show).pack()
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

animals = ["Tý","Sửu","Dần","Mão","Thìn","Tỵ",
           "Ngọ","Mùi","Thân","Dậu","Tuất","Hợi"]

def convert():
    y = int(e.get())
    can = (y - 4) % 10
    chi = (y - 4) % 12
    messagebox.showinfo("Kết quả",
        f"Năm âm lịch: {animals[chi]}")

root = tk.Tk()
tk.Label(root, text="Năm dương lịch").pack()
e = tk.Entry(root)
e.pack()
tk.Button(root, text="Chuyển đổi", command=convert).pack()
root.mainloop()
```

```python
import tkinter as tk

def bmi():
    w = float(e1.get())
    h = float(e2.get())
    b = w/(h*h)
    kq = "Gầy" if b<18.5 else "Bình thường" if b<25 else "Thừa cân"
    lbl.config(text=f"BMI: {b:.2f}\nKết luận: {kq}")

root = tk.Tk()
tk.Label(root, text="Cân nặng (kg)").pack()
e1 = tk.Entry(root); e1.pack()
tk.Label(root, text="Chiều cao (m)").pack()
e2 = tk.Entry(root); e2.pack()
tk.Button(root, text="Tính BMI", command=bmi).pack()
lbl = tk.Label(root)
lbl.pack()
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def show():
    messagebox.showinfo("Lập trình hướng đối tượng",
                        "Viết phương trình giải phương trình bậc 2")

root = tk.Tk()
menu = tk.Menu(root)
root.config(menu=menu)

chuong1 = tk.Menu(menu, tearoff=0)
menu.add_cascade(label="Chương 1", menu=chuong1)
chuong1.add_command(label="1.1", command=show)
chuong1.add_separator()
chuong1.add_command(label="Exit", command=root.quit)

root.mainloop()
```

```python
import tkinter as tk

def taxi():
    km = float(e1.get())
    wait = int(e2.get())
    money = 12000

    if km > 0.8:
        if km <= 30:
            money += (km-0.8)*15300
        else:
            money += (30-0.8)*15300 + (km-30)*12100

    if wait > 5:
        money += (wait-5)*750

    lbl.config(text=f"Tiền taxi: {int(money):,} đ")

root = tk.Tk()
tk.Label(root, text="Số km").pack()
e1 = tk.Entry(root); e1.pack()
tk.Label(root, text="Thời gian chờ (phút)").pack()
e2 = tk.Entry(root); e2.pack()
tk.Button(root, text="Tính tiền", command=taxi).pack()
lbl = tk.Label(root)
lbl.pack()
root.mainloop()
```
