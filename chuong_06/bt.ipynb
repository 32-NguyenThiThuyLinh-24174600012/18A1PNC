```python
import threading

def show():
    print("Luồng đang chạy:", threading.current_thread().name)

threads = []
for i in range(3):
    t = threading.Thread(target=show, name=f"Thread-{i+1}")
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import threading, urllib.request

urls = [
    "https://www.example.com",
    "https://www.python.org",
]

def download(url):
    name = url.split("//")[1].split(".")[0]
    urllib.request.urlretrieve(url, name + ".html")
    print("Đã tải:", url)

threads = []
for u in urls:
    t = threading.Thread(target=download, args=(u,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import threading

def even():
    for i in range(30, 51):
        if i % 2 == 0:
            print("Chẵn:", i)

def odd():
    for i in range(30, 51):
        if i % 2 != 0:
            print("Lẻ:", i)

t1 = threading.Thread(target=even)
t2 = threading.Thread(target=odd)

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
import threading
import math

def factorial(n):
    print(f"Giai thừa {n} =", math.factorial(n))

nums = [5, 6, 7]
threads = []

for n in nums:
    t = threading.Thread(target=factorial, args=(n,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import threading, requests

urls = [
    "https://www.google.com",
    "https://www.python.org"
]

def fetch(url):
    r = requests.get(url)
    print(url, "->", r.status_code)

threads = []
for u in urls:
    t = threading.Thread(target=fetch, args=(u,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
import threading

def even(n):
    for i in range(0, n+1, 2):
        print("Chẵn:", i)

def odd(n):
    for i in range(1, n+1, 2):
        print("Lẻ:", i)

n = 20
t1 = threading.Thread(target=even, args=(n,))
t2 = threading.Thread(target=odd, args=(n,))

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
import threading, time

def web(name, delay, count):
    print(f"Starting {name}")
    for _ in range(count):
        time.sleep(delay)
        print(f"{name}: {time.ctime()}")
    print(f"Exiting {name}")

print("Starting Main Thread")

t1 = threading.Thread(target=web, args=("Google",1,4))
t2 = threading.Thread(target=web, args=("Yahoo",2,3))
t3 = threading.Thread(target=web, args=("Facebook",3,3))

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()

print("Exiting Main Thread")
```

```python
import threading

data = [5,4,7,4,6,2,1,7,7,7,4,7,8,7,2,2,4,5,6,0,2,5,8,8,5,4,7,3,4,1,0,0,9,6,5,0,3,9]
num_threads = 3

total_sum = 0
lock = threading.Lock()

def part_sum(part, idx):
    global total_sum
    s = sum(part)
    print(f"Thread {idx}: " + "; ".join(map(str, part)) + ";")
    with lock:
        total_sum += s

size = len(data) // num_threads
threads = []

for i in range(num_threads):
    part = data[i*size:] if i == num_threads-1 else data[i*size:(i+1)*size]
    t = threading.Thread(target=part_sum, args=(part, i+1))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("Tổng list =", total_sum)
```

```python
import threading, random

data = [random.randint(0,100) for _ in range(30)]
num_threads = 3
max_values = []

lock = threading.Lock()

def find_max(part):
    with lock:
        max_values.append(max(part))

size = len(data) // num_threads
threads = []

for i in range(num_threads):
    part = data[i*size:] if i == num_threads-1 else data[i*size:(i+1)*size]
    t = threading.Thread(target=find_max, args=(part,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("List:", data)
print("Giá trị lớn nhất =", max(max_values))
```
