```python
import json

# T·∫°o m·ªôt ƒë·ªëi t∆∞·ª£ng t·ª´ ƒëi·ªÉn Python
person = {
    "name": "Nguyen Van A",
    "age": 25,
    "city": "Hanoi",
    "email": "vana@example.com",
    "skills": ["Python", "Machine Learning", "Data Analysis"]
}

# Chuy·ªÉn ƒë·ªïi dict th√†nh chu·ªói JSON (s·∫Øp x·∫øp theo kh√≥a, th·ª•t l·ªÅ 4)
json_string = json.dumps(person, indent=4, sort_keys=True)

# In chu·ªói JSON ƒë·∫πp m·∫Øt
print("‚úÖ ƒê·ªëi t∆∞·ª£ng JSON sau khi chuy·ªÉn ƒë·ªïi v√† s·∫Øp x·∫øp:")
print(json_string)

# N·∫øu mu·ªën in t·ª´ng gi√° tr·ªã ri√™ng l·∫ª (c√°c th√†nh vi√™n ƒë·ªëi t∆∞·ª£ng):
print("\nüìã C√°c gi√° tr·ªã trong ƒë·ªëi t∆∞·ª£ng JSON:")
# Chuy·ªÉn l·∫°i t·ª´ JSON string sang dict ƒë·ªÉ truy c·∫≠p
json_obj = json.loads(json_string)

for key, value in json_obj.items():
    print(f"{key}: {value}")
```
