```python
import json

# Đọc dữ liệu từ file JSON
with open("data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Lấy thông tin công ty
company_name = data["company"]["name"]
address = data["company"]["address"]
departments = data["company"]["departments"]

# Tính tổng số nhân viên
total_employees = sum(len(dept["employees"]) for dept in departments)

# In thông tin công ty
print(f"Tên công ty: {company_name}")
print(f"Địa chỉ: {address}")
print(f"Tổng số nhân viên: {total_employees}\n")

print("----- Thống kê nhân viên theo đơn vị -----\n")

# Duyệt qua từng đơn vị
for idx, dept in enumerate(departments, start=1):
    dept_name = dept["name"]
    emp_count = len(dept["employees"])
    ratio = (emp_count / total_employees * 100) if total_employees > 0 else 0
    print(f"{idx}. Tên đơn vị: {dept_name}")
    print(f"- Số nhân viên: {emp_count}")
    print(f"- Tỷ lệ: {ratio:.2f}%\n")
```
