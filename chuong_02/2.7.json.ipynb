```python
import json

# Chu盻擁 JSON (gi蘯｣ s盻ｭ nh蘯ｭn t盻ｫ API ho蘯ｷc file)
json_data = '''
{
    "name": "Nguyen Van A",
    "age": 25,
    "city": "Hanoi",
    "skills": ["Python", "Machine Learning", "Data Analysis"]
}
'''

# Chuy盻ハ ﾄ黛ｻ品 JSON thﾃnh ﾄ黛ｻ訴 tﾆｰ盻｣ng Python (dict)
python_obj = json.loads(json_data)

# In ra k蘯ｿt qu蘯｣
print("笨 ﾄ雪ｻ訴 tﾆｰ盻｣ng Python sau khi chuy盻ハ ﾄ黛ｻ品:")
print(python_obj)

# Truy c蘯ｭp d盻ｯ li盻u nhﾆｰ dictionary
print("\n沒 Truy c蘯ｭp d盻ｯ li盻u:")
print("Tﾃｪn:", python_obj["name"])
print("Tu盻品:", python_obj["age"])
print("K盻ｹ nﾄハg:", ", ".join(python_obj["skills"]))
```
