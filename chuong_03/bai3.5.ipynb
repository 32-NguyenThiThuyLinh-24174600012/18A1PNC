```python
import numpy as np
with open("heights_1.txt") as f:
    height = [float(x) for x in f.read().split(",")]

with open("weights_1.txt") as f:
    weight = [float(x) for x in f.read().split(",")]
arr_height = np.array(height)
arr_weight = np.array(weight)
# Đổi đơn vị
arr_height_m = arr_height * 0.0254
arr_weight_kg = arr_weight * 0.453592
# BMI
arr_bmi = arr_weight_kg / (arr_height_m ** 2)
n = arr_height_m.size
print("Số phần tử:", n)
if n > 50:
    print("Index 50 -> height:", arr_height_m[50], "m; weight:", arr_weight_kg[50], "kg")
else:
    print(f"Không có index 50 (chỉ có 0..{n-1}). In phần tử cuối thay thế:")
    print("Index", n-1, "-> height:", arr_height_m[-1], "m; weight:", arr_weight_kg[-1], "kg")
print("Các phần tử từ 100 đến 110:", arr_height_m[100:111])
print("BMI < 21:", arr_bmi[arr_bmi < 21])
# Trung bình & lớn nhất/nhỏ nhất
print("Chiều cao TB:", np.mean(arr_height_m))
print("Cân nặng TB:", np.mean(arr_weight_kg))
print("BMI TB:", np.mean(arr_bmi))
print("Chiều cao max/min:", np.max(arr_height_m), "/", np.min(arr_height_m))
print("Cân nặng max/min:", np.max(arr_weight_kg), "/", np.min(arr_weight_kg))
```
