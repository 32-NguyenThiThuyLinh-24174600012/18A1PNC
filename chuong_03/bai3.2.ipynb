```python
import numpy as np

arr = np.arange(10)
print("Mảng arr:", arr)
print("Kiểu dữ liệu:", arr.dtype)
print("Kích thước:", arr.shape)
```

```python
arr_odd = arr[arr % 2 == 1]
arr_even = arr[arr % 2 == 0]

print("arr_odd:", arr_odd)
print("arr_even:", arr_even)
```

```python
arr_update_1 = np.where(arr % 2 == 0, arr, arr + 10)
print("arr_update_1:", arr_update_1)
```
