```python
import pandas as pd

# 1 Đọc file stocks1.csv
stocks1 = pd.read_csv("stocks1.csv")
print("Stocks1 - 5 dòng đầu:")
print(stocks1.head())
print("\nStocks1 - 5 dòng cuối:")
print(stocks1.tail())
print("\nKiểu dữ liệu columns:")
print(stocks1.dtypes)
print("\nThông tin stocks1:")
print(stocks1.info())

# 1 b) Đọc file stocks2.csv
stocks2 = pd.read_csv("stocks2.csv")
print("\nStocks2 - 5 dòng đầu:")
print(stocks2.head())
print("\nStocks2 - 5 dòng cuối:")
print(stocks2.tail())
print("\nKiểu dữ liệu columns:")
print(stocks2.dtypes)
print("\nThông tin stocks2:")
print(stocks2.info())

# 1 c) Đọc file companies.csv
companies = pd.read_csv("companies.csv")
print("\nCompanies:")
print(companies.head())
print("\nKiểu dữ liệu columns:")
print(companies.dtypes)
print("\nThông tin companies:")
print(companies.info())

# 2 Kiểm tra dữ liệu Null trong stocks1
print("\nSố giá trị Null trong stocks1:")
print(stocks1.isnull().sum())

# Thay thế Null
# Nếu Null cột 'high' -> max theo symbol
stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max()))
# Nếu Null cột 'low' -> min theo symbol
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min()))

# 3 Gộp stocks1 và stocks2 theo dòng
stocks = pd.concat([stocks1, stocks2], ignore_index=True)
print("\n15 dòng cuối stocks:")
print(stocks.tail(15))

# 4 Gộp stocks và companies
# Giả sử 'symbol' trong stocks tương ứng với 'name' trong companies
stocks_companies = stocks.merge(companies, left_on='symbol', right_on='name', how='left')
print("\n5 dòng đầu stocks_companies:")
print(stocks_companies.head())

# 5 Giá trung bình và volume trung bình của mỗi công ty
mean_stats = stocks_companies.groupby('symbol')[['open','high','low','close','volume']].mean()
print("\nGiá và volume trung bình mỗi công ty:")
print(mean_stats)

# 6 Giá đóng cửa (close) trung bình, lớn nhất, nhỏ nhất
close_stats = stocks_companies.groupby('symbol')['close'].agg(['mean','max','min'])
print("\nGiá đóng cửa trung bình, max, min mỗi công ty:")
print(close_stats)

# 7 Tạo cột parsed_time dạng DateTime
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
print("\nKiểu dữ liệu cột parsed_time:")
print(stocks_companies['parsed_time'].dtypes)
print("\n5 dòng đầu stocks_companies:")
print(stocks_companies.head())

# 8 Thêm cột result
stocks_companies['result'] = stocks_companies.apply(
    lambda row: 'up' if row['close'] > row['open'] else 'down', axis=1
)
print("\n5 dòng đầu stocks_companies với cột result:")
print(stocks_companies.head())
```
