```python
import numpy as np

# Đọc dữ liệu an toàn
baseball = []
with open("baseball_2D.txt", "r") as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) != 2:
            continue  # bỏ qua dòng không hợp lệ
        h, w = parts
        baseball.append([float(h), float(w)]) 

# Tạo numpy array
np_baseball = np.array(baseball)

print("Kiểu dữ liệu:", np_baseball.dtype)
print("Kích thước (shape):", np_baseball.shape)

# In dòng thứ 5 (vì chỉ có 20 cầu thủ)
print("Dòng thứ 5:", np_baseball[4])

# Tạo mảng cân nặng
np_weight = np_baseball[:, 1]
print("Mảng cân nặng:", np_weight)

# Chiều cao của vận động viên thứ 10
height_10 = np_baseball[9, 0]
print("Chiều cao của vận động viên thứ 10:", height_10, "inch")

# Trung bình chiều cao và cân nặng
mean_height = np.mean(np_baseball[:, 0])
mean_weight = np.mean(np_baseball[:, 1])
print("Chiều cao trung bình:", mean_height)
print("Cân nặng trung bình:", mean_weight)

# Hệ số tương quan
corr = np.corrcoef(np_baseball[:, 0], np_baseball[:, 1])[0, 1]
print("Hệ số tương quan giữa chiều cao và cân nặng:", corr)

# Nhận xét
if corr > 0.5:
    print("→ Có tương quan thuận mạnh giữa chiều cao và cân nặng.")
elif corr > 0:
    print("→ Có tương quan thuận yếu giữa chiều cao và cân nặng.")
elif corr < -0.5:
    print("→ Có tương quan nghịch mạnh.")
else:
    print("→ Hầu như không có tương quan.")
```
