```python
import pandas as pd

# Đọc dữ liệu từ file CSV
euro12 = pd.read_csv("euro2012.csv")

# 0 Thông tin tổng quát
print("Kiểu dữ liệu của euro12:", type(euro12))
print("Kích thước (shape):", euro12.shape)
print("Danh sách các cột:")
print(euro12.columns)

# 1 In giá trị cột 'Goals'
print("\n1. Cột Goals:")
print(euro12['Goals'])

# 2 Số đội tham gia Euro 2012
num_teams = euro12['Team'].nunique()
print("\n2. Số đội tham gia Euro 2012:", num_teams)

# 3 Thông tin tổng quát của DataFrame
print("\n3. Thông tin Euro2012:")
print(euro12.info())

# 4 Tạo DataFrame 'discipline'
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\n4. DataFrame discipline:")
print(discipline.head())

# 5 Sắp xếp giảm dần theo Red Cards rồi Yellow Cards
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)
print("\n5. discipline sắp xếp giảm dần:")
print(discipline_sorted)

# 6 a) Trung bình thẻ vàng
mean_yellow = discipline['Yellow Cards'].mean()
print("\n6a. Trung bình thẻ vàng:", mean_yellow)

# 6 b) Lọc đội có hơn 6 bàn thắng
more_than_6_goals = euro12[euro12['Goals'] > 6]
print("\n6b. Các đội có hơn 6 bàn thắng:")
print(more_than_6_goals[['Team', 'Goals']])

# 7 Đội bắt đầu bằng chữ 'G'
teams_G = euro12[euro12['Team'].str.startswith('G')]
print("\n7. Các đội bắt đầu bằng 'G':")
print(teams_G[['Team', 'Goals']])

# 8 In 7 cột đầu tiên
print("\n8. 7 cột đầu:")
print(euro12.iloc[:, :7].head())

# 9 In tất cả trừ 3 cột cuối
print("\n9. Tất cả trừ 3 cột cuối:")
print(euro12.iloc[:, :-3].head())

#10 In các cột: Team, Goals, Shooting Accuracy, Yellow Cards, Red Cards
print("\n10. Các cột được chọn:")
print(euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']].head())

# 11 In 'Team' và 'Shooting Accuracy' cho England, Italy, Russia
selected_teams = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])]
print("\n11. Shooting Accuracy của England, Italy, Russia:")
print(selected_teams[['Team', 'Shooting Accuracy']])
```
