```python
import pandas as pd

# 1 Đọc dữ liệu từ file drinks.csv với index_col=0
drink = pd.read_csv("drinks.csv", index_col=0)

# Thông tin cơ bản
print("Kiểu dữ liệu:", type(drink))
print("Kích thước (shape):", drink.shape)
print("Tên các cột:", drink.columns)
print("\n5 dòng đầu:")
print(drink.head())
print("\n5 dòng cuối:")
print(drink.tail())

# 2 Số lượng bia tiêu thụ trung bình theo châu lục
mean_beer_by_continent = drink.groupby('continent')['beer_servings'].mean()
print("\n2. Trung bình tiêu thụ bia theo châu lục:")
print(mean_beer_by_continent)

# 3 Thống kê số lượng rượu vang theo châu lục
wine_stats_by_continent = drink.groupby('continent')['wine_servings'].describe()
print("\n3. Thống kê số lượng rượu vang theo châu lục:")
print(wine_stats_by_continent)

# 4 Trung bình số bia và rượu tiêu thụ theo châu lục
mean_drinks_by_continent = drink.groupby('continent')[['beer_servings', 'wine_servings']].mean()
print("\n4. Trung bình bia và rượu theo châu lục:")
print(mean_drinks_by_continent)

# 5 Trung vị (median) số bia và rượu theo châu lục
median_drinks_by_continent = drink.groupby('continent')[['beer_servings', 'wine_servings']].median()
print("\n5. Trung vị bia và rượu theo châu lục:")
print(median_drinks_by_continent)

# 6 Thống kê rượu mạnh (spirit_servings) theo châu lục
spirit_stats = drink.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min'])
print("\n6. Rượu mạnh theo châu lục (mean, max, min):")
print(spirit_stats)

# 7 Sắp xếp theo số bia tiêu thụ
drink_sorted = drink.sort_values(by='beer_servings')
print("\n5 quốc gia tiêu thụ bia ít nhất:")
print(drink_sorted.head())
print("\n5 quốc gia tiêu thụ bia nhiều nhất:")
print(drink_sorted.tail())
```
