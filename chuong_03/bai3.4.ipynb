```python
import numpy as np

# 1️⃣
arr_zeros = np.zeros(10)
arr_zeros[6] = 11
print("arr_zeros:", arr_zeros)

# 2️⃣
arr_h = np.arange(10, 25)[::-1]
print("arr_h:", arr_h)

# 3️⃣
arr_k = np.array([1, 2, 0, 8, 2, 3, 5, 0])
arr_l = arr_k[arr_k != 0]
print("arr_l:", arr_l)

# 4️⃣
arr_l = np.append(arr_l, [10, 20])
print("arr_l sau khi thêm:", arr_l)

# 5️⃣
arr_l = np.insert(arr_l, 5, [100, 101, 102])
print("arr_l sau khi chèn:", arr_l)

# 6️⃣
arr_l = np.delete(arr_l, [0, 1, 2])
print("arr_l sau khi xóa:", arr_l)
```
