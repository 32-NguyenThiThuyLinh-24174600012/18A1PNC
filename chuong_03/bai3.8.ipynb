```python
import numpy as np

# Đọc dữ liệu từ file
heights = []
positions = []

with open("heights.txt", "r") as f:
    for line in f:
        heights.append(float(line.strip()))

with open("positions.txt", "r") as f:
    for line in f:
        positions.append(line.strip())

# 1a. Tạo numpy array
np_positions = np.array(positions)
np_heights = np.array(heights)

print("Kiểu dữ liệu np_positions:", np_positions.dtype)
print("Kiểu dữ liệu np_heights:", np_heights.dtype)

# 2. Tính chiều cao trung bình của các GK
mean_GK = np.mean(np_heights[np_positions == "GK"])
print("Chiều cao trung bình của GK:", mean_GK)

# 3. Chiều cao trung bình của vị trí khác GK
mean_others = np.mean(np_heights[np_positions != "GK"])
print("Chiều cao trung bình của các vị trí khác GK:", mean_others)

# 4. Tạo structured array (mảng có cấu trúc)
players = np.array(
    list(zip(np_positions, np_heights)),
    dtype=[('position', 'U5'), ('height', 'f4')]
)

# 5. Sắp xếp theo chiều cao
players_sorted = np.sort(players, order='height')

print("\n--- Dữ liệu sắp xếp theo chiều cao ---")
print(players_sorted)

# Lấy vị trí có chiều cao cao nhất và thấp nhất
tallest = players_sorted[-1]
shortest = players_sorted[0]

print("\nCầu thủ cao nhất:", tallest['position'], "-", tallest['height'])
print("Cầu thủ thấp nhất:", shortest['position'], "-", shortest['height'])
```
