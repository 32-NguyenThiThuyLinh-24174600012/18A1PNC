```python
class ThiSinh:
    def __init__(self):
        self.hoten = ""
        self.toan = 0.0
        self.ly = 0.0
        self.hoa = 0.0
    
    def nhap(self):
        self.hoten = input("Nhập họ tên thí sinh: ")
        self.toan = float(input("Điểm Toán: "))
        self.ly = float(input("Điểm Lý: "))
        self.hoa = float(input("Điểm Hóa: "))
    
    def tong_diem(self):
        return self.toan + self.ly + self.hoa
    
    def in_thong_tin(self):
        print(f"{self.hoten} - Toán: {self.toan}, Lý: {self.ly}, Hóa: {self.hoa}, Tổng điểm: {self.tong_diem()}")

# Nhập danh sách thí sinh
n = int(input("Nhập số thí sinh: "))
ds_thi_sinh = []

for _ in range(n):
    ts = ThiSinh()
    ts.nhap()
    ds_thi_sinh.append(ts)

# Lọc thí sinh trúng tuyển (tổng điểm >= 20) và sắp xếp giảm dần theo tổng điểm
ds_trung_tuyen = [ts for ts in ds_thi_sinh if ts.tong_diem() >= 20]
ds_trung_tuyen.sort(key=lambda x: x.tong_diem(), reverse=True)

print("\nDanh sách thí sinh trúng tuyển:")
for ts in ds_trung_tuyen:
    ts.in_thong_tin()
```

```python
class PS:
    def __init__(self):
        self.tu = 0
        self.mau = 1
    
    def hop_le(self):
        return self.mau != 0
    
    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        while not self.hop_le():
            print("Mẫu số không hợp lệ (phải khác 0). Nhập lại.")
            self.mau = int(input("Nhập mẫu số: "))
    
    def in_phan_so(self):
        print(f"{self.tu}/{self.mau}")

# Ví dụ sử dụng
p = PS()
p.nhap()
print("Phân số vừa nhập là:", end=" ")
p.in_phan_so()
```

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.data = []
    
    def isEmpty(self):
        return len(self.data) == 0
    
    def isFull(self):
        return len(self.data) == self.size
    
    def push(self, value):
        if self.isFull():
            print("Ngăn xếp đã đầy, không thể push.")
        else:
            self.data.append(float(value))
    
    def pop(self):
        if self.isEmpty():
            print("Ngăn xếp rỗng, không thể pop.")
            return None
        else:
            return self.data.pop()
    
    def __del__(self):
        # Hàm huỷ (tùy chọn, ở Python thường không cần)
        self.data.clear()
        print("Ngăn xếp đã được huỷ.")

# Ví dụ sử dụng
stack = Stack(3)
stack.push(1.1)
stack.push(2.2)
stack.push(3.3)
stack.push(4.4)  # Thông báo ngăn xếp đầy

print(stack.pop())  # 3.3
print(stack.pop())  # 2.2
print(stack.pop())  # 1.1
print(stack.pop())  # Thông báo rỗng
```

```python
class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year
    
    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}")
    
    def is_leap_year(self):
        # Kiểm tra năm nhuận
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)
    
    def days_in_month(self):
        # Số ngày trong tháng
        if self.month in [1,3,5,7,8,10,12]:
            return 31
        elif self.month in [4,6,9,11]:
            return 30
        elif self.month == 2:
            return 29 if self.is_leap_year() else 28
        else:
            return 0  # tháng không hợp lệ
    
    def next(self):
        max_day = self.days_in_month()
        if max_day == 0:
            print("Ngày không hợp lệ")
            return
        
        if self.day < max_day:
            self.day += 1
        else:
            self.day = 1
            if self.month < 12:
                self.month += 1
            else:
                self.month = 1
                self.year += 1

# Ví dụ sử dụng:
d = Date(28, 2, 2024)
d.display()  # 28/02/2024
d.next()
d.display()  # 29/02/2024 (năm nhuận)

d.next()
d.display()  # 01/03/2024
```

```python
import math

class DaGiac:
    def __init__(self, so_canh=3, canh=1):
        self.so_canh = so_canh
        self.canh = canh  # Giả sử đa giác đều, mỗi cạnh bằng nhau
    
    def chu_vi(self):
        return self.so_canh * self.canh
    
    def dien_tich(self):
        # Diện tích đa giác đều: (n * s^2) / (4 * tan(pi/n))
        if self.so_canh < 3:
            return 0
        return (self.so_canh * self.canh ** 2) / (4 * math.tan(math.pi / self.so_canh))
    
    def in_thong_tin(self):
        print(f"Đa giác đều {self.so_canh} cạnh, mỗi cạnh = {self.canh}")
        print(f"Chu vi: {self.chu_vi():.2f}")
        print(f"Diện tích: {self.dien_tich():.2f}")

class HinhBinhHanh(DaGiac):
    def __init__(self, canh_day, chieu_cao, canh_xuong):
        super().__init__(so_canh=4)
        self.canh_day = canh_day  # đáy
        self.chieu_cao = chieu_cao
        self.canh_xuong = canh_xuong  # cạnh bên
    
    def chu_vi(self):
        return 2 * (self.canh_day + self.canh_xuong)
    
    def dien_tich(self):
        return self.canh_day * self.chieu_cao
    
    def in_thong_tin(self):
        print(f"Hình bình hành: đáy = {self.canh_day}, cạnh bên = {self.canh_xuong}, chiều cao = {self.chieu_cao}")
        print(f"Chu vi: {self.chu_vi():.2f}")
        print(f"Diện tích: {self.dien_tich():.2f}")

class HinhChuNhat(HinhBinhHanh):
    def __init__(self, dai, rong):
        super().__init__(canh_day=dai, chieu_cao=rong, canh_xuong=rong)
    
    # Chu vi và diện tích kế thừa từ Hình bình hành là đúng
    
    def in_thong_tin(self):
        print(f"Hình chữ nhật: dài = {self.canh_day}, rộng = {self.chieu_cao}")
        print(f"Chu vi: {self.chu_vi():.2f}")
        print(f"Diện tích: {self.dien_tich():.2f}")

class HinhVuong(HinhChuNhat):
    def __init__(self, canh):
        super().__init__(dai=canh, rong=canh)
    
    def in_thong_tin(self):
        print(f"Hình vuông: cạnh = {self.canh_day}")
        print(f"Chu vi: {self.chu_vi():.2f}")
        print(f"Diện tích: {self.dien_tich():.2f}")

# Ví dụ sử dụng:

print("Đa giác đều:")
dg = DaGiac(5, 3)
dg.in_thong_tin()

print("\nHình bình hành:")
hb = HinhBinhHanh(canh_day=5, chieu_cao=3, canh_xuong=4)
hb.in_thong_tin()

print("\nHình chữ nhật:")
hcn = HinhChuNhat(dai=6, rong=3)
hcn.in_thong_tin()

print("\nHình vuông:")
hv = HinhVuong(canh=4)
hv.in_thong_tin()
```

```python
import math

# Lớp điểm
class Diem:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def nhap(self):
        self.x = float(input("Nhập hoành độ x: "))
        self.y = float(input("Nhập tung độ y: "))
    
    def display(self):
        print(f"Tâm: ({self.x}, {self.y})")

# Lớp Elip kế thừa từ Diem
class Elip(Diem):
    def __init__(self, x=0, y=0, a=1, b=1):
        super().__init__(x, y)
        self.a = a  # Bán trục lớn
        self.b = b  # Bán trục nhỏ

    def nhap(self):
        super().nhap()
        self.a = float(input("Nhập bán trục lớn a: "))
        self.b = float(input("Nhập bán trục nhỏ b: "))

    def dien_tich(self):
        return math.pi * self.a * self.b

    def display(self):
        super().display()
        print(f"Bán trục lớn a = {self.a}, bán trục nhỏ b = {self.b}")
        print(f"Diện tích elip: {self.dien_tich():.2f}")

# Lớp Đường tròn kế thừa từ Elip (vì đường tròn là elip đặc biệt với a = b = r)
class DuongTron(Elip):
    def __init__(self, x=0, y=0, r=1):
        super().__init__(x, y, r, r)  # a = b = r

    def nhap(self):
        super().nhap()
        self.a = self.b = float(input("Nhập bán kính r: "))

    def display(self):
        print(f"Tâm đường tròn: ({self.x}, {self.y})")
        print(f"Bán kính r = {self.a}")
        print(f"Diện tích đường tròn: {self.dien_tich():.2f}")
print("=== Nhập Elip ===")
elip = Elip()
elip.nhap()
elip.display()

print("\n=== Nhập Đường tròn ===")
circle = DuongTron()
circle.nhap()
circle.display()
```

```python
import math

class TamGiac:
    def __init__(self, a=1, b=1, c=1):
        self.a = a
        self.b = b
        self.c = c

    def nhap(self):
        self.a = float(input("Nhập cạnh a: "))
        self.b = float(input("Nhập cạnh b: "))
        self.c = float(input("Nhập cạnh c: "))

    def la_tam_giac(self):
        return self.a + self.b > self.c and \
               self.a + self.c > self.b and \
               self.b + self.c > self.a

    def chu_vi(self):
        return self.a + self.b + self.c

    def dien_tich(self):
        if not self.la_tam_giac():
            return 0
        p = self.chu_vi() / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))

    def display(self):
        print(f"Các cạnh: a = {self.a}, b = {self.b}, c = {self.c}")
        if self.la_tam_giac():
            print(f"Chu vi: {self.chu_vi():.2f}")
            print(f"Diện tích: {self.dien_tich():.2f}")
        else:
            print("Không phải là tam giác hợp lệ.")

# Tam giác vuông
class TamGiacVuong(TamGiac):
    def la_tam_giac_vuong(self):
        canh = sorted([self.a, self.b, self.c])
        return math.isclose(canh[0]**2 + canh[1]**2, canh[2]**2)

    def display(self):
        super().display()
        if self.la_tam_giac() and self.la_tam_giac_vuong():
            print("=> Đây là tam giác vuông.")
        elif self.la_tam_giac():
            print("=> Không phải tam giác vuông.")

# Tam giác cân
class TamGiacCan(TamGiac):
    def la_tam_giac_can(self):
        return self.a == self.b or self.a == self.c or self.b == self.c

    def display(self):
        super().display()
        if self.la_tam_giac() and self.la_tam_giac_can():
            print("=> Đây là tam giác cân.")
        elif self.la_tam_giac():
            print("=> Không phải tam giác cân.")

# Tam giác đều kế thừa từ tam giác cân
class TamGiacDeu(TamGiacCan):
    def la_tam_giac_deu(self):
        return self.a == self.b == self.c

    def display(self):
        super().display()
        if self.la_tam_giac() and self.la_tam_giac_deu():
            print("=> Đây là tam giác đều.")
        elif self.la_tam_giac():
            print("=> Không phải tam giác đều.")
print("=== Tam giác thường ===")
tg = TamGiac()
tg.nhap()
tg.display()

print("\n=== Tam giác vuông ===")
tg_vuong = TamGiacVuong()
tg_vuong.nhap()
tg_vuong.display()

print("\n=== Tam giác cân ===")
tg_can = TamGiacCan()
tg_can.nhap()
tg_can.display()

print("\n=== Tam giác đều ===")
tg_deu = TamGiacDeu()
tg_deu.nhap()
tg_deu.display()

```
